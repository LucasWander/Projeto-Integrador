```python
#Importar bibliotecas
import requests
import geopandas as gpd
import rtree
import pygeos
import mapclassify
import geobr
import folium
import pandas as pd
```

```python
geojson_file = "/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/POLÍGONO_3_BAIRROS/setores-3-bairros-Rio_Claro.geojson"
```

```python
#Plotar 3 bairros
m = folium.Map([-22.410821233296186, -47.55966967960197], zoom_start=12)

gdf = gpd.read_file(geojson_file)

#folium.GeoJson(gdf,).add_to(m)
folium.GeoJson(geojson_file).add_to(m)

#Outras opções de tiles (leaflet-extras.github.io/leaflet-providers/preview)
folium.TileLayer(tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
                 attr = 'Tiles &copy; Esri &mdash; Source: Esri, i-cubed, USDA, USGS, AEX, GeoEye, Getmapping, Aerogrid, IGN, IGP, UPR-EGP, and the GIS User Community',
                 name = "Esri.WorldImagery").add_to(m)

#Adicionando controle de camadas
folium.LayerControl().add_to(m)

m
```
