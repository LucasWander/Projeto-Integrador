```python
#Importar bibliotecas
import requests
import geopandas as gpd
import rtree
import pygeos
import mapclassify
import geobr
import folium
import pandas as pd
```

```python
geojson_file = "/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/POLÍGONO_3_BAIRROS/setores-3-bairros-Rio_Claro.geojson"
```

```python
gdf = gpd.read_file(geojson_file)
```

```python
gdf.head()
```

```python
#Filtrando o bairro Floridiana
floridiana = gdf[gdf.NM_BAIRRO=='Jardim Floridiana']
floridiana
```

```python
#Salvar arquivo .shp
# Caminho onde o arquivo .shp será salvo
caminho_saida = "/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/POLÍGONO_3_BAIRROS/floridiana.shp"

# Salvando o GeoDataFrame como um arquivo .shp
floridiana.to_file(caminho_saida)

floridiana_geojson_file = "/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/POLÍGONO_3_BAIRROS/floridiana.geojson"

# Salvando como .geojson
floridiana.to_file(floridiana_geojson_file, driver='GeoJSON')
```

```python
#Filtrando o bairro Grande Cervezão
chervezon = gdf[gdf.NM_BAIRRO=='Grande Cervezão']
chervezon
```

```python
#Salvar arquivo .shp
# Caminho onde o arquivo .shp será salvo
caminho_saida = "/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/POLÍGONO_3_BAIRROS/chervezon.shp"

# Salvando o GeoDataFrame como um arquivo .shp
chervezon.to_file(caminho_saida)

chervezon_geojson_file = "/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/POLÍGONO_3_BAIRROS/chervezon.geojson"

# Salvando como .geojson
chervezon.to_file(chervezon_geojson_file, driver='GeoJSON')
```

```python
#Filtrando o bairro Centro
centro = gdf[gdf.NM_BAIRRO=='Centro']
centro
```

```python
#Salvar arquivo .shp
# Caminho onde o arquivo .shp será salvo
caminho_saida = "/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/POLÍGONO_3_BAIRROS/centro.shp"

# Salvando o GeoDataFrame como um arquivo .shp
centro.to_file(caminho_saida)

centro_geojson_file = "/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/POLÍGONO_3_BAIRROS/centro.geojson"

# Salvando como .geojson
centro.to_file(centro_geojson_file, driver='GeoJSON')
```

```python
#Plotar 3 bairros
m = folium.Map([-22.410821233296186, -47.55966967960197], zoom_start=12)

#folium.GeoJson(gdf,).add_to(m)
folium.GeoJson(floridiana_geojson_file).add_to(m)
folium.GeoJson(chervezon_geojson_file).add_to(m)
folium.GeoJson(centro_geojson_file).add_to(m)

#Outras opções de tiles (leaflet-extras.github.io/leaflet-providers/preview)
folium.TileLayer(tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
                 attr = 'Tiles &copy; Esri &mdash; Source: Esri, i-cubed, USDA, USGS, AEX, GeoEye, Getmapping, Aerogrid, IGN, IGP, UPR-EGP, and the GIS User Community',
                 name = "Esri.WorldImagery").add_to(m)

#Adicionando controle de camadas
folium.LayerControl().add_to(m)

m.save('/content/drive/MyDrive/Colab Notebooks/UNIVESP/PI/PYTHON_MAPAS/VERSÃO_FINAL/mapa_3_bairros.html')

m
```
